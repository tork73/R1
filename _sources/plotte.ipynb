# Plotte med Python

Vi kan plotte diagrammer og grafer med Python. Da bruker vi ofte modulen matplotlib.pyplot. Vi imorterer denne og lager et alias (en forkortelse) slik: 

```python
import matplotlib.pyplot as plt
```

Vi viser hvordan vi kan gj√∏re de ulike plottene gjennom en del eksempler. 

### Eksempel 1

Tabellen nedenfor viser temperaturen for noen klokkeslett. 

| Kokkeslett | 10 | 12 | 13   | 14   | 15   | 16   | 17   |
|------------|:----:|:----:|:------:|:------:|:------:|:------:|:------:|
| Temperatur $^\circ C$| 12 | 13 | 14.3 | 14.5 | 15.3 | 14.2 | 15.0 |

Plott tallen i tabellen. 


```python
# Vi lager to lister, en for x-verdier og en for y-verdier:

Klokkeslett = [10, 12, 13, 14, 15, 16, 17]
Temperatur = [12, 13, 14.3, 14.5, 15.3, 14.2, 15]

plt.plot(Klokkeslett, Temperatur, "-*")
plt.xlabel('Klokkeslett')
plt.ylabel('Temperatur')
plt.show()
```

## Plotting av kontinuerlige funksjoner

Dersom vi skal plotte kontinuerlige funksjoner, m√• vi lage mange x-verdier som vi kan bruke. For √• lage y-verdien definerer vi funksjonen og finner y-verdiene tilh√∏rende hver x-verdi. Dersom vi bruker numpy er dette ganske greit √• gj√∏re. 

### Eksempel 3

Plott grafen til $f(x)=x^2-2x-4$ i et koordinatsystem. Velg x-verdier mellom $-3$ og $5$. 

```python
import matplotlib.pyplot as plt # For √• kunne plotte
import numpy as np              # For √• kunne lage en x-verdier og y-verdier

# Definerer funksjonen:

def f(x):
    return x**2-2*x-4

# definerer x-verdier og y-verdier

X = np.linspace(-3, 5, 100)  # 100 x-verdier mellom -3 og 5
Y = f(X)                     # 100 y-verdier 

plt.plot(X, Y)               # Plotter x-verdier og y-verdier
plt.xlabel("x")              # Legger til navn p√• x-aksen
plt.ylabel("y")              # Legger til navn p√• y-aksen
plt.title(f"Grafen til $f(x)=x^2-2x-4$") # Legger til tittel
plt.grid()             # Legger til rutenett p√• grafen
plt.show()
```

### Eksempel 4

Funksjonen $f$ er gitt ved

$$f(x) = x ^ 5+x ^ 3, \ \ D_f = [-1, \ 1] $$

Tegn grafen til $f$ sammen med grafen til den omvendte funksjonen $f ^ {-1}$ i
samme  koordinatsystem.


```python
import numpy as np
import matplotlib.pyplot as plt 

X = np.linspace(-1, 1, 100)
def f(x):
    return x**5+x**3

Y = f(X)
plt.style.use("seaborn")
plt.plot(X, Y, label=r"$f$")
plt.plot(Y, X, label=r"$f^{-1}$")
plt.plot([-2, 2], [-2, 2],"--", label=r"$y=x$")
plt.hlines(0, -2, 2)
plt.vlines(0, -2, 2)
plt.title(r"Grafen til $f$ og $f^{-1}$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
